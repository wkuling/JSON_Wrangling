## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas-docs.github.io/pandas-docs-travis/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# Excercise 3: In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Unfortunately, isnull applied to an empty dictionary gives back a false, so I need a lambda function...
f=lambda x: len(str(x)) > 0
json_filter=json_norm['name'].map(f)
```

```python
# Let's look at what element 1 looks like
json_filtered=json_norm[json_filter]
```

```python
json_filtered
```

```python
json_filtered=json_filtered.groupby(['code', 'name']).size()
```

```python
json_filtered
```

```python
json_filtered['1']
```

```python
df= pd.DataFrame(json_filtered)
df
```

```python
df=df.reset_index(['name'])
#df.drop(['0'], axis=1, inplace=True)
```

```python
df
```

```python
df.drop([0], axis=1, inplace=True)
```

```python
df
```

```python
type(json_norm)
```

```python
json_norm=json_norm.merge(df.reset_index(['code']), on='code', how='left')
```

```python
json_norm=json_norm.drop('name_x', axis=1)
```

```python
json_norm.rename(columns={'name_y': 'name'}, inplace=True)
```

```python
json_norm
```

```python
# OK, so that worked...
```

```python
json_filled=json_norm.replace(json_filtered)
json_filled
```
